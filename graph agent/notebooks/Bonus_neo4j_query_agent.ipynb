````python
from agno.agent import Agent
from agno.models.groq import Groq
from agno.tools.neo4j import Neo4jTools

blockstat_agent = Agent(
    name="BlockStat Forensic Graph Analyst",
    model=Groq(id="openai/gpt-oss-120b"),  # YOUR MODEL! 🎯
    tools=[Neo4jTools()],
    instructions=[
        "BlockStat forensic analyst for blockchain fraud detection.",
        "SINGLE Cypher statements only. Limit 1-2 tool calls per response.",
        "Schema: Wallet, Mixer, Token | HOLDS, TRANSFER relationships.",
        "USDC: 0xA0b86991c6218b36c1d19D4a2e9Eb0cE3606eB48",
        "Output: Risk Score | Top Whales | Mixer Flows | Suspicious Clusters",
    ],
    markdown=True
)

print("🔍 BlockStat Forensic Agent READY!")
print("✅ Using openai/gpt-oss-120b ")
````

````python
from neo4j import GraphDatabase

driver = GraphDatabase.driver(
    os.getenv('NEO4J_URI'),
    auth=(os.getenv('NEO4J_USERNAME'), os.getenv('NEO4J_PASSWORD'))
)

with driver.session(database=os.getenv('NEO4J_DATABASE')) as session:
    result = session.run("RETURN 'Neo4j AuraDB Connected!' AS message").single()
    print(f"✅ {result['message']}")
````

````python
def load_forensic_data():
    with driver.session(database=os.getenv('NEO4J_DATABASE')) as session:
        # 1. Clear database
        session.run("MATCH (n) DETACH DELETE n")
        print("🗑️ Cleared database")

        # 2. Create USDC Token
        session.run("""
            CREATE (t:Token {address: '0xA0b86991c6218b36c1d19D4a2e9Eb0cE3606eB48',
                          symbol: 'USDC', name: 'USD Coin'})
        """)
        print("✅ USDC Token created")

        # 3. Create 50 Wallets ONLY (SIMPLE)
        session.run("""
            UNWIND range(0,49) AS i
            CREATE (w:Wallet {address: '0xWALLET' + toString(i),
                           balance: 1000 + i*10,
                           type: CASE WHEN i < 10 THEN 'Mixer'
                                  WHEN i < 15 THEN 'WashTrader'
                                  ELSE 'Holder' END,
                           risk: CASE WHEN i < 15 THEN 'high' ELSE 'low' END})
        """)
        print("✅ 50 Wallets created")

        # 4. Create HOLDS relationships (SIMPLE)
        session.run("""
            MATCH (t:Token {address: '0xA0b86991c6218b36c1d19D4a2e9Eb0cE3606eB48'})
            MATCH (w:Wallet)
            CREATE (w)-[:HOLDS {balance: w.balance}]->(t)
        """)
        print("✅ 50 HOLDS created")

        # 5. Add Mixer labels (SIMPLE)
        session.run("""
            MATCH (w:Wallet {type: 'Mixer'})
            SET w:Mixer
        """)
        print("✅ 10 Mixer labels added")

        # 6. Create 200 Transfers (SIMPLE - NO complex CASE)
        session.run("""
            UNWIND range(0,199) AS idx
            WITH idx, idx % 50 AS srcIdx, (idx*7 % 50) AS dstIdx
            WHERE srcIdx <> dstIdx
            MATCH (src:Wallet {address: '0xWALLET' + toString(srcIdx)})
            MATCH (dst:Wallet {address: '0xWALLET' + toString(dstIdx)})
            CREATE (src)-[:TRANSFER {
                amount: toFloat(10 + (idx % 90)),
                risk: 'low',
                txHash: '0x' + substring('0123456789abcdef', idx % 16, 8)
            }]->(dst)
        """)
        print("✅ 200 Transfers created")

        # 7. Verify (SIMPLE counts)
        nodes = session.run("MATCH (n) RETURN count(n) AS nodes").single()['nodes']
        rels = session.run("MATCH ()-[r]->() RETURN count(r) AS rels").single()['rels']
        mixers = session.run("MATCH (m:Mixer) RETURN count(m) AS mixers").single()['mixers']

        print(f"✅ LOADED: {nodes} nodes, {rels} rels, {mixers} mixers!")

load_forensic_data()
print("🎉 BlockStat USDC forensic data PERFECTLY LOADED!")
````
